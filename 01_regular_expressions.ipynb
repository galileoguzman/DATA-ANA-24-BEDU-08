```python
import re
import pandas as pd
```

```python
# Cargar nuestro dataset
df = pd.read_csv('tmp/Ejemplo_01_03_Amazon_Sales_Dataset.csv')

df.head()
```

### Filtrar de numeros de telefono que comiencen con 101

```python
def find_phone_number_with_101(text):
    pattern = r'\b101-\d{3}-\d{4}' # 101-903-9622
    result = re.findall(pattern, text) # Buscame el patron/coincidencias (pattern) en el objeto text
    return bool(result)

df_with_101_phone_numbers = df[df['Phone Number'].apply(find_phone_number_with_101)]
df_with_101_phone_numbers.head()
```

#### Filtrar de resenas con la palabra 'excellent'

```python
def find_reviews_with_excellent(text):
    pattern = r'\bexcellent\b|\bExcellent\b'
    result = re.findall(pattern, text)
    return bool(result)

df_reviews_with_excellent = df[df['Review'].apply(find_reviews_with_excellent)]
df_reviews_with_excellent.head()
```

### Busqueda con search de productos que comiencen con 'Tablet'

```python
def find_products_starts_with_tablet(text):
    pattern = r'\bTablet\b'
    result = re.search(pattern, text)
    return bool(result)

df_products_starts_with_tablet = df[df['Product Name'].apply(find_products_starts_with_tablet)]
df_products_starts_with_tablet.head()
```

```python
def find_name_with_garcia(text):
    pattern = r'\bGarcia\b'
    result = re.search(pattern, text, re.IGNORECASE)
    return bool(result)

df_name_with_garcia = df[df['Full Name'].apply(find_name_with_garcia)]
df_name_with_garcia.head()
```

#### Ejemplos de regex con split

```python
def split_pricing(text):
    parts = re.split(r'\.', str(text))
    if len(parts) == 2:
        return parts[0], parts[1] # 211.77 -> (211, 77)
    return parts[0], '00' # 299 -> (299, 00)

df[['Price integer part', 'Price decimal part']] = df['Price'].apply(lambda x: pd.Series(split_pricing(x)))
df.head()
```
